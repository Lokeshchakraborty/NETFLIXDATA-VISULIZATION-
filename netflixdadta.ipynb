```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv('netflix_titles.csv')
df
```

```python
#data cleqning
df=df.dropna(subset=['type','release_year','rating','country','duration'])
type_counts=df['type'].value_counts()
type_counts
```

```python
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=['red','blue'])
plt.title('number of movies ns tv shows on netflix')
plt.xlabel('type')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('movies vs tv shows ')
plt.show()
```

```python
unique_values = df['rating'].unique()
print(unique_values)
```

```python
rating=df['rating'].value_counts()
print(rating)
rating_type=['PG-13' ,'TV-MA', 'TV-14' ,'TV-Y7' ,'PG', 'R' ,'TV-PG', 'TV-Y', 'TV-G', 'G',
 'NC-17', 'NR' ,'TV-Y7-FV' ,'UR']
plt.pie(rating,labels=rating_type,autopct='%1.1f%%',startangle=90)
plt.title('% of content rating ')
plt.tight_layout()
plt.savefig('content rating pie  ')
plt.show()
```

```python
movie_df=df[df['type']=='Movie'].copy()
movie_df['duration_int']=movie_df['duration'].str.replace(' min','').astype(int)
plt.figure(figsize=(8,6))
plt.hist(movie_df['duration_int'],bins=30,color='red',edgecolor='black')
plt.title('duration histogram ')
plt.xlabel('duration')
plt.ylabel('count')

plt.tight_layout()
plt.savefig('duration histogram ')
plt.show()
```

```python
release_counts=df['release_year'].value_counts().sort_index()
plt.figure(figsize=(8,6))
plt.scatter(release_counts.index,release_counts.values,color='red')
plt.title('release_year vs count ')
plt.xlabel('release_year')
plt.ylabel('count')

plt.tight_layout()
plt.savefig('release counts ')
plt.show()
```

```python
country_counts=df['country'].value_counts().head(10)
plt.figure(figsize=(8,6))
plt.barh(country_counts.index,country_counts.values,color=['red'])
plt.title('top 10 countries by numver of shows ')
plt.xlabel('numver of shows')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('top 10 countries by numver of shows ')

plt.show()
```

```python
content_by_year=df.groupby(['release_year','type']).size().unstack().fillna(0)

fig,ax=plt.subplots(1,2,figsize=(12,5))

ax[0].plot(content_by_year.index,content_by_year['Movie'],color='red')
ax[0].set_title('Top 10 movies by year')
ax[0].set_xlabel('year')
ax[0].set_ylabel('count')


ax[0].plot(content_by_year.index,content_by_year['TV Show'],color='yellow')
ax[0].set_title('tv shows by year')
ax[0].set_xlabel('year')
ax[0].set_ylabel('count')

fig.suptitle('comparion of movies and tv shows by year')
plt.tight_layout()
plt.savefig('top 10 movies by year')
plt.show()
```
